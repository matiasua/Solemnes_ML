## Separacion y escalado del dataset.

```python
y = df['quality']
X = df.drop('quality', axis=1)
```

```python
X
```

```python
#Escalar estandar.
scaler = StandardScaler()


#Ajusto el escalar basandome en los valores de X, luego transformo X y lo almacenamos nuevamente en el como un dataframe.
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

## BÃºsqueda del cluster con el dendrograma

```python
plt.figure(figsize=(15, 10))
plt.title("Dendrograma")
dendrograma = shc.dendrogram(shc.linkage(X, method = 'ward'))
```

### Segun lo que logro interpretar desde el dendrograma, la cantidad de clusters que interpreto es de 5 o 4.

### Sabemos que existen 6 clasificaciones de vinos, por lo cual el dendrograma no detecto uno o dos clusters adicionales. (Esto a criterio de mi interpretacion)

### La verdad me cuesta bastante interpretar el dendrograma.


### Haremos la comparacion y visualizacion de los agrupamientos entre 5 y  4 clusters.


# Agrupamiento y Visualizacion

```python
#defino el algoritmo con 5 clusters
model = AgglomerativeClustering(n_clusters = 5, 
                                    affinity='euclidean', linkage='ward')
```

```python
model.fit(X)
```

### Utilizare una tecnica llamada PCA (Principal Component Analysis).
### reduce la cantidad de variables a visualizar, creando una cantidad menor de nuevas variables que representen lo mejor posible a las variables originales.

```python
pca = PCA(n_components=2)

X_reducido = pd.DataFrame(pca.fit_transform(X), columns = ['PC1', 'PC2'])
X_reducido
```

```python
targetdf = pd.DataFrame()
targetdf['target'] = model.labels_
df_final = pd.concat([X_reducido, targetdf], axis = 1)
```

```python
fig = plt.figure(figsize = (7,5))
aux = fig.add_subplot(1,1,1) 
aux.set_xlabel('', fontsize = 15)
aux.set_ylabel('', fontsize = 15)
aux.set_title('5 clusters', fontsize = 20)
targets = [0,1,2,3,4]
colors = ['slateblue','springgreen','indigo','teal','lightcoral']
for target, color in zip(targets,colors):
    indicesToKeep = df_final['target'] == target
    aux.scatter(df_final.loc[indicesToKeep, 'PC1'], df_final.loc[indicesToKeep, 'PC2'], c = color, s = 50)
aux.legend(targets)
aux.grid()
```

```python
#Se define el algoritmo con 4 clusters
model2 = AgglomerativeClustering(n_clusters = 4, 
                                    affinity='euclidean', linkage='ward')
```

```python
#Se entrena el algoritmo
model2.fit(X)
```

```python
pca = PCA(n_components=2)

X_reducido = pd.DataFrame(pca.fit_transform(X), columns = ['PC1', 'PC2'])
X_reducido
```

```python
targetdf = pd.DataFrame()
targetdf['target'] = model2.labels_
df_final = pd.concat([X_reducido, targetdf], axis = 1)
```

```python
fig = plt.figure(figsize = (7,5))
aux = fig.add_subplot(1,1,1) 
aux.set_xlabel('', fontsize = 15)
aux.set_ylabel('', fontsize = 15)
aux.set_title('4 clusters', fontsize = 20)
targets = [0,1,2,3]
colors = ['slateblue','springgreen','indigo','teal']
for target, color in zip(targets,colors):
    indicesToKeep = df_final['target'] == target
    aux.scatter(df_final.loc[indicesToKeep, 'PC1'], df_final.loc[indicesToKeep, 'PC2'], c = color, s = 50)
aux.legend(targets)
aux.grid()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html
model = MLPClassifier(hidden_layer_sizes=(256, 256), max_iter=500)

model.fit(X_train, y_train)
```

```python
print(f"Precision del modelo: {model.score(X_test, y_test)}")
```
